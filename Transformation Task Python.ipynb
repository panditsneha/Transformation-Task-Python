### Importing libraries

### Reading input files

```python
ip = pd.read_csv('Sample Input - Sample.csv.csv')
comments = op = pd.read_csv('comments.csv')
```

### Concatenating Date and Time column and restoring it in the Date column

```python
time_stamp = ip['Date'].astype(str) +' '+ip['Time']
ip['Date']= time_stamp
ip.drop(columns=['Time','Milli Sec'],inplace=True)
ip.head()
```

### Grouping Date column for similar 'datetime' and taking maximum from all the columns

```python
ip = ip.groupby('Date',as_index=False).max()
ip.head()
```

### Converting the time interval into 10 seconds and taking the desired values

```python
dff = ip.drop('Date',axis=1)
out = pd.DataFrame()
date_time = []
for i in range(0,64,10):
    df = dff.loc[i:i+9]
    date_time.append(ip['Date'][i])
    out = out.append(df.max(),ignore_index=True)
colm = list(dff.columns)[0:7]
all_list = [out[colm[i]][i] for i in range(7)]
```

### Creating a new dataframe to store all the required columns

```python
data = {'event_timestamp': date_time,
        'tag__name': colm[:7],
        'tag__value': all_list}
data = pd.DataFrame(data)
```

```python
data.insert(1, 'tag_id', comments['ItemId'].loc[:6])
data.insert(3, 'tag_desc', comments['Comment'].loc[:6])
data['tag__unit'] = data['tag_desc'].str.extract(r"\(([A-Za-z]+)\)", expand=False)
```
